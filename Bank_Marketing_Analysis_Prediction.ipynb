# Data Preprocessing

```python
train_input=bank.drop(['age','pdays','default'],axis=1, inplace=True)
```

Encoding classified attributes

```python
job = pd.get_dummies(bank['job'],drop_first=True)
marital = pd.get_dummies(bank['marital'],drop_first=True)
education = pd.get_dummies(bank['education'],drop_first=True)
#default = pd.get_dummies(bank['default'],drop_first=True)
housing = pd.get_dummies(bank['housing'],drop_first=True)
loan = pd.get_dummies(bank['loan'],drop_first=True)
contact = pd.get_dummies(bank['contact'],drop_first=True)
month = pd.get_dummies(bank['month'],drop_first=True)
poutcome = pd.get_dummies(bank['poutcome'],drop_first=True)
```

```python
train_input=bank.drop(['job','marital','education','housing','loan','contact','month','poutcome','y'],axis=1)
train_input = pd.concat([train_input,job,marital,education,housing,loan, contact,month,poutcome],axis=1)
```

Training Data

```python
train_input.info()
```

```python
train_output=bank['y']
```

```python
train_output.head()
```

```python
train_input.head()
```

Test data

```python
bank_test.head()
```

```python
test_input=bank_test.drop(['age','pdays','default'],axis=1, inplace=True)
```

```python
job_test = pd.get_dummies(bank_test['job'],drop_first=True)
marital_test = pd.get_dummies(bank_test['marital'],drop_first=True)
education_test = pd.get_dummies(bank_test['education'],drop_first=True)
#default_test = pd.get_dummies(bank_test['default'],drop_first=True)
housing_test = pd.get_dummies(bank_test['housing'],drop_first=True)
loan_test = pd.get_dummies(bank_test['loan'],drop_first=True)
contact_test = pd.get_dummies(bank_test['contact'],drop_first=True)
month_test = pd.get_dummies(bank_test['month'],drop_first=True)
poutcome_test = pd.get_dummies(bank_test['poutcome'],drop_first=True)
```

```python
test_input=bank_test.drop(['job','marital','education','housing','loan','contact','month','poutcome','y'],axis=1)
test_input = pd.concat([test_input,job_test,marital_test,education_test,housing_test,loan_test,contact_test,month_test,poutcome_test],axis=1)
```

```python
test_input.info()
```

```python
test_output=bank_test['y']
```

# Decision Tree Model and Prediction

```python
from sklearn.tree import DecisionTreeClassifier  
dtree = DecisionTreeClassifier()  
dtree.fit(train_input, train_output)  
```

```python
dtree_pred=dtree.predict(test_input)
```

```python
dtree_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix  
print(confusion_matrix(test_output, dtree_pred))  
print(classification_report(test_output, dtree_pred))  
```

# Random Forest Model and Prediction

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()  
rfc.fit(train_input, train_output)
```

```python
rfc_pred=rfc.predict(test_input)
```

```python
rfc_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix  
print(confusion_matrix(test_output, rfc_pred))  
print(classification_report(test_output, rfc_pred))
```

# Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(train_input,train_output)
```

```python
log_pred=log.predict(test_input)
```

```python
log_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix  
print(confusion_matrix(test_output, log_pred))  
print(classification_report(test_output, log_pred))
```

# K-Nearest Neighbor

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(train_input,train_output)
```

```python
knn_pred=knn.predict(test_input)
```

```python
knn_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix  
print(confusion_matrix(test_output, knn_pred))  
print(classification_report(test_output, knn_pred))
```
